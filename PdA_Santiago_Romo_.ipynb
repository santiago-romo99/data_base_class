```python
# Trabajo de Acompa√±amiento: MEDICAMENTOS VITALES NO DISPONIBLES
print("Trabajo de Acompa√±amiento: MEDICAMENTOS VITALES NO DISPONIBLES")
print("Se selecciono una amplia base de datos que recopila informacion sobre medicamentos vitales no disponibles en Colombia, esta brinda informacion como nombre del principo activo del medicamento, precio, marca, concentracion, unidad de dispensacion,etc :")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
url_csv = 'https://www.datos.gov.co/resource/3t73-n4q9.csv'
df_online = pd.read_csv(url_csv)
print(df_online.head())

# Top 10 precio por unidad
print("Acontinuacion se presenta una grafica de barras donde se puede ver el top de precios de medicamentos de esta base de datos, se puede ver que el mas caro doblega el segundo, brindandonos una perspectiva de como los precios de los medicamentos ocilan demasiado")
df_top10 = df_online.sort_values(by='precio_por_tableta', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_top10, x='precio_por_tableta', y='principio_activo', palette='viridis')

plt.title("Top 10 precios por tableta")
plt.xlabel("Precio por tableta")
plt.ylabel("principio activo")

plt.tight_layout()
plt.show()


# Unidad de dispensacion

print("El siguiente grafico muestra la que unidades de dispensacion son las mas feceuntes de estos medicamentos vitales, se puede analizar que las tabletas son las mas recurrentes, esto tambien indica que la via oral es la mas utilizada para la dministracion de medicamentos ")
df_unidades = df_online['unidad_de_dispensacion'].value_counts().reset_index()
df_unidades.columns = ['unidad_de_dispensacion', 'cantidad']
plt.figure(figsize=(10, 6))
sns.barplot(data=df_unidades,
            x='unidad_de_dispensacion',
            y='cantidad',
            palette='Set2')

plt.title("Cantidad de repeticiones por unidad de dispensaci√≥n")
plt.xlabel("Unidad de dispensaci√≥n")
plt.ylabel("Cantidad")
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

# Mayores fabricantes
top3_fabricantes = df_online['fabricante'].value_counts().head(3)

print("Top 3 fabricantes m√°s repetidos:")
print("En esta parte se estima cuales son los mayores fabricantes de medicamentos teniendo en cuanta la fecuencia en la que aparecen por cada medicamento reportado, indicando que marcas y compa√±ias tienen el monopolio farmaceutico en Colombia")
print(top3_fabricantes)
```
