```python
# 2
matriz = np.random.randint(10, 51, size=(4, 4))
matriz_modificada = np.where(matriz % 2 != 0, -1, matriz)
print(matriz_modificada)
```

```python
#3
matriz = np.random.rand(5, 5)
matriz_normal = (matriz- matriz.min()) / (matriz.max() - matriz.min())
print("Matriz 5x5 con valores entre 0 y 1:")
print(matriz_normal)
suma_columnas = np.sum(matriz_normal, axis=0)
print("\nSuma por columnas:")
print(suma_columnas)
```

```python
#4
arr = np.array([1, 2, 3, 4, 5])
arr_invertido = arr[::-1]
print("Array invertido:", arr_invertido)
```

```python
#5
identidad = np.identity(6)
matriz_resultado = 3 * identidad
print(matriz_resultado)

#El resultado es una matriz diagonal donde todos los elementos de la diagonal son 3 debido a que se multiplica 3x1 de la matriz identida
```

```python
#6
array = np.random.randint(1, 101, size=15)
array_ordenado = np.sort(array)[::-1]
print("Array ordenado de mayor a menor:", array_ordenado)
```

```python
#7
array = np.random.randint(0, 101, size=30)
media = np.mean(array)
mediana = np.median(array)
desviacion = np.std(array)
print("Array:", array)
print("Media:", media)
print("Mediana:", mediana)
print("Desviación estándar:", desviacion)
```

```python
#8
matriz = np.random.randint(0, 101, size=(6, 3))
promedios_filas = np.mean(matriz, axis=1)
indice_menor_promedio = np.argmin(promedios_filas)
matriz[indice_menor_promedio] = 0
print(matriz)
```

```python
#9
matriz = np.random.randint(1, 11, size=(3, 4))
cumsum_filas = np.cumsum(matriz, axis=1)
print("Matriz original:")
print(matriz)
print("Suma por filas:")
print(cumsum_filas)
```

```python
#punto 3

df7 = pd.DataFrame({
'Producto': ['A', 'B', 'A', 'C', 'B', 'A'],
'Precio': [1200, 1500, 1300, 1000, 1600, 1250],
'Cantidad': [3, 2, 5, 1, 2, 4]})

df7['Ingreso'] = df7['Precio'] * df7['Cantidad']
print(df7)

ingresos_por_producto = df7.groupby('Producto')['Ingreso'].sum().reset_index()
print(ingresos_por_producto )

ingresos_ordenados = ingresos_por_producto.sort_values(by='Ingreso', ascending=False)
print(ingresos_ordenados)
```

```python
np.random.seed(0)
productos = ['Arroz', 'Azúcar', 'Aceite', 'Café', 'Sal', 'Harina', 'Lentejas']
zonas = ['Norte', 'Sur', 'Este', 'Oeste', 'Centro', 'Noroccidente']

df10 = pd.DataFrame({
    'producto': np.random.choice(productos, size=1000),
    'precio': np.random.randint(1000, 6000, size=1000),
    'cantidad': np.random.randint(1, 11, size=1000),
    'dia': np.random.randint(1, 30, size=1000),
    'zona': np.random.choice(zonas, size=1000)})


ventas_por_zona = df10.groupby(['zona', 'producto'])['cantidad'].sum().reset_index()
producto_mas_vendido = ventas_por_zona.sort_values(['zona', 'cantidad'], ascending=[True, False])\
                                       .drop_duplicates(subset='zona', keep='first')
print("producto más vendido por zona:")
print(producto_mas_vendido)

```

```python
 ingresos_por_dia = df10.groupby('dia').apply(lambda x: (x['precio'] * x['cantidad']).sum()).reset_index(name='ganancia')
 print(ingresos_por_dia)
```

```python
dia_max = ingresos_por_dia.loc[ingresos_por_dia['ganancia'].idxmax()]
print(dia_max)
```
