```python
# Solución Pregunta 1
import matplotlib.pyplot as plt

meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]
temperaturas = [22, 24, 27, 30, 32, 35, 34, 33, 30, 27, 24, 22]

plt.plot(meses, temperaturas, color='red', marker='o')
plt.title("Temperatura media mensual")
plt.xlabel("Mes")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.show()
```

```python
# Solución Pregunta 2
import seaborn as sns
import pandas as pd
import numpy as np

np.random.seed(0)
df2 = pd.DataFrame({
    'Edad': np.random.randint(18, 60, 20),
    'Ingreso': np.random.randint(1000, 5000, 20),
    'Genero': np.random.choice(['Masculino', 'Femenino'], 20)
})
sns.set_style("whitegrid")
sns.scatterplot(data=df2, x='Edad', y='Ingreso', hue='Genero', s=100)
plt.title("Relación entre Edad e Ingreso mensual")
plt.show()
```

```python
# Solución Pregunta 3
datos = np.random.normal(50, 10, 200)
plt.hist(datos, bins=15, color='blue', alpha=0.6)
plt.title("Distribución de datos normales")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Solución Pregunta 4
df4 = pd.DataFrame({
    'Altura': np.concatenate([
        np.random.normal(1.3, 0.1, 30),
        np.random.normal(1.7, 0.1, 30),
        np.random.normal(1.65, 0.08, 30)
    ]),
    'Grupo': ['Niños']*30 + ['Adultos']*30 + ['Adultos Mayores']*30
})
sns.set_style("whitegrid")
sns.boxplot(data=df4, x='Grupo', y='Altura', palette='Set2', showfliers=True)
plt.title("Comparación de alturas por grupo")
plt.show()
```

```python
# Solución Pregunta 5
meses = list(range(1, 13))
regiones = ['Norte', 'Centro', 'Sur']

df5 = pd.DataFrame({
    'Mes': np.tile(meses, len(regiones)),
    'Region': np.repeat(regiones, len(meses)),
    'Ventas': np.random.randint(1000, 5000, len(meses)*len(regiones))
})
ventas_totales = df5.groupby('Mes')['Ventas'].sum().reset_index()
ventas_totales['Promedio_movil'] = ventas_totales['Ventas'].rolling(window=3, min_periods=1).mean()
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(data=df5, x='Mes', y='Ventas', hue='Region', palette='Set2', ci=None)
plt.plot(ventas_totales['Mes'], ventas_totales['Promedio_movil'],
         color='black', marker='o', linewidth=2, label='Promedio móvil')
plt.title("Ventas mensuales por región con promedio móvil")
plt.ylabel("Ventas")
plt.xlabel("Mes")
plt.legend(title="", labels=["Promedio móvil", "Norte", "Centro", "Sur"], loc='upper left')

plt.tight_layout()
plt.show()
```
