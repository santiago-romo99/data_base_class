```python
# 13. filtrar con multiples condicionales

operatorOr = df1[(df1['Altura'] < 1.50) | (df1['Nombre_Completo'] != 'Elena')]
operatorAnd = df1[(df1['Altura'] < 1.50) & (df1['Nombre_Completo'] != 'Elena')]

print(operatorOr)
print(operatorAnd)


```

```python
operatorOr = df2[(df2['Peso_corporal'] < 70) | (df2['Nombre_Completo'] != 'Daniela')]
operatorAnd = df2[(df2['Peso_corporal'] < 70) & (df2['Nombre_Completo'] != 'Daniela')]

print(operatorOr)
print(operatorAnd)


```

```python
operatorOr = df3[(df3['Indice_Masa_Corporal'] < 24) | (df3['Nombre_Completo'] != 'Gael')]
operatorAnd = df3[(df3['Indice_Masa_Corporal'] < 24) & (df3['Nombre_Completo'] != 'Gael')]

print(operatorOr)
print(operatorAnd)
```
